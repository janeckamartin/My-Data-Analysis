```python
# !pip install dash

import dash
from dash import dcc
from dash import html

from dash.dependencies import Input, Output

import plotly.express as px

import threading
from google.colab import output


df = px.data.iris()
species = df['species'].unique()

app = dash.Dash()

elements = []

left = []
right = []

elements.append(html.H1('Hello Dash', id='h1'))

left.append(html.H3('Choose species:', id='h3'))


dropdown_species = dcc.Dropdown(
    options=species,
    value=species[0],
    id='dropdown_species',
    clearable=False
)
left.append(dropdown_species)
left.append(dcc.Graph(id='graph'))

div = html.Div(left, id='div', style={'display':'inline-block', 'width': '48%', 'height':600, "background-color": 'gray', "margin-right": '4%'})

elements.append(div)


right.append(html.H3('Choose species:', id='h3b'))

dropdown_species = dcc.Dropdown(
    options=species,
    value=species[0],
    id='dropdown_species2',
    clearable=False
)
right.append(dropdown_species)
right.append(dcc.Graph(id='graph2'))

div = html.Div(right, id='div', style={'display':'inline-block', 'width': '48%', 'height':600, "background-color": 'gray'})

elements.append(div)



@app.callback(
    Output('graph', 'figure'),
    Input('dropdown_species', 'value')
)
def update_graph(species):
  d_species = {
      'setosa': 'red',
      'versicolor': 'green',
      'virginica': 'blue'
  }

  df_species = df[df['species'] == species]
  figure = px.scatter(df_species, x='sepal_length', y='sepal_width', color_discrete_sequence=[d_species[species]])

  return figure

@app.callback(
    Output('graph2', 'figure'),
    Input('dropdown_species2', 'value')
)
def update_graph(species):
  d_species = {
      'setosa': 'red',
      'versicolor': 'green',
      'virginica': 'blue'
  }

  df_species = df[df['species'] == species]
  figure = px.scatter(df_species, x='petal_length', y='petal_width', color_discrete_sequence=[d_species[species]])

  return figure


app.layout = html.Div(elements)

thread = threading.Thread(target=app.run)
thread.start()
output.serve_kernel_port_as_iframe(port=8050, height='600')
```
