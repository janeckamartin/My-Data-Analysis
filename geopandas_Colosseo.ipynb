```python
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
import zipfile
with zipfile.ZipFile("colosseo.zip","r") as zip_ref:
    zip_ref.extractall("colosseo")
```

```python
import geopandas as gpd
```

```python
colosseo_natural = gpd.read_file("/content/colosseo/colosseo/shape/natural.shp")
colosseo_buildings = gpd.read_file("/content/colosseo/colosseo/shape/buildings.shp")
colosseo_roads = gpd.read_file("/content/colosseo/colosseo/shape/roads.shp")
colosseo_railways = gpd.read_file('/content/colosseo/colosseo/shape/railways.shp')
colosseo_landuse = gpd.read_file('/content/colosseo/colosseo/shape/landuse.shp')

my_places = gpd.read_file('map.geojson')

colosseo_natural.head()
```

```python
colosseo_natural.to_crs(colosseo_roads.crs)
colosseo_buildings.to_crs(colosseo_roads.crs)
colosseo_railways.to_crs(colosseo_roads.crs)
colosseo_landuse.to_crs(colosseo_roads.crs)
colosseo_roads.to_crs(colosseo_roads.crs)

fix, ax = plt.subplots(figsize=(16, 14))

my_places.plot(ax=ax, color="#ff8a5b")

colosseo_natural.plot(ax=ax, color="#BB9BB0")
colosseo_buildings.plot(ax=ax, color="#CCBCBC")
colosseo_roads.plot(ax=ax, color="#1C1D21")
colosseo_railways.plot(ax=ax, color="#B61634", alpha=0.7)
colosseo_landuse.plot(ax=ax, color="#4F7CAC")

plt.axis('off')
plt.show()
```
