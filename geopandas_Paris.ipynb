```python
import geopandas as gpd
import matplotlib.pyplot as plt

import zipfile
with zipfile.ZipFile("Paris.zip","r") as zip_ref:
    zip_ref.extractall("Paris")

```

```python
roads = gpd.read_file("/content/Paris/Paris/shape/roads.shp")
roads_others = roads[roads["maxspeed"] != 30]
roads_30 = roads[roads["maxspeed"] == 30.0]

fig, ax = plt.subplots(figsize=(10, 10))

roads_others.plot(ax=ax, color="black", alpha=0.5)
roads_30.plot(ax=ax, color="red", alpha=0.5)

plt.axis('off')

plt.show()
```
