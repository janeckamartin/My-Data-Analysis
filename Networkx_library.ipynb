```python
import matplotlib.pyplot as plt
import networkx as nx

plt.axis('off')

G = nx.DiGraph()

edge_list = [('B', 'A'), ('C', 'B'), ('B', 'E'), ('A', 'C'),('C', 'D'),('D', 'E')]
G.add_edges_from(edge_list)

pos={'B': (0, 0), 'A': (1, 3), 'C': (2, 1), 'D': (4, 2), 'E': (4, 0)}

ax=plt.gca()
ax.margins(0.1)

nx.draw_networkx(G, pos=pos, with_labels=True)
```
