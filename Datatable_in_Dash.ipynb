```python
!pip install dash

import dash
from dash import dcc
from dash import html
from dash import dash_table

from dash.dash_table import DataTable
from dash.dash_table.Format import Format, Scheme, Symbol

import pandas as pd

from dash.dependencies import Input, Output

import plotly.express as px

import threading
from google.colab import output


df = px.data.iris()

app = dash.Dash()

number_format = Format(
  precision=1,
  scheme=Scheme.fixed,
  symbol=Symbol.yes,
  symbol_suffix=u' cm'
)

app.layout = dash_table.DataTable(
    id='table',
    columns=[
        {"name": ["sepal", "length"], "id": "sepal_length", 'type': 'numeric', 'format': number_format},
        {"name": ["sepal", "width"], "id": "sepal_width", 'type': 'numeric'},
        {"name": ["petal", "length"], "id": "petal_length", 'type': 'numeric'},
        {"name": ["petal", "width"], "id": "petal_width", 'type': 'numeric'},
        {"name":["",'species'], 'id':'species', 'type': 'text'},
        {"name":["","species_id"],'id':'species_id', 'type': 'numeric'}],
    data=df.to_dict('records'),
    page_size=10,
    style_header={
        'backgroundColor': 'white',
        'fontWeight': 'bold'
    },
    merge_duplicate_headers=True,
    # style_as_list_view=True,

    style_data_conditional=[
    {
        'if': {'row_index': 'odd'},
        'backgroundColor': 'rgb(220, 220, 220)',

    },
    {
        'if': {'column_type': 'text'},
        'textAlign': 'left'
    },
    {
        'if': {'column_type': 'text'},
        'textIndent': 10
    },
    ]
)

thread = threading.Thread(target=app.run)
thread.start()
output.serve_kernel_port_as_iframe(port=8050, height='600')














```
